```python
# Renaming columns
def rename_columns(Data, old_col_names,new_col_names):
    """
    Renames columns in the dataframe
    arguments:
          Data:the dataframe to rename columns,
          old_col_names: names of the old columns
          new_col_names: names of the new renamed columns

    return Data : returns dataframe with renamed columns
    """
    if len(old_col_names) != len(new_col_names):
        raise ValueError("Number of old columns should be equal to number of new columns")
    mapping={old:new for old, new in zip(old_col_names, new_col_names)}
    Data.rename(columns=mapping, inplace=True)
    return Data
old_col_names=[ 'how_many_children_do_you_have__if_any',
       'Have you ever had health insurance?', 'If yes, which insurance cover?',
       'When was the last time you visited a hospital for medical treatment? (In Months)',
       'Did you have health insurance during your last hospital visit?',
       'Have you ever had a routine check-up with a doctor or healthcare provider?',
       'If you answered yes to the previous question, what time period (in years) do you stay before having your routine check-up?',
       'Have you ever had a cancer screening (e.g. mammogram, colonoscopy, etc.)?',
       'If you answered yes to the previous question, what time period (in years) do you stay before having your Cancer screening?']

new_col_names=['Children','Had_Insurance','Insurance_Company','Last_time_visited_hospital(months)',
               'Had_insurance_in_hospital','Has_checkup_with_doctor',
               'routine_checkup_period(years)','Cancer_screening','Cancer_screening_period(years)']
Data=rename_columns(Data,old_col_names,new_col_names)
```

```python
# Dropping Unnecessary columns
columns_to_drop=['Your Picture', 'Your Picture_URL', '_id', '_uuid', '_submission_time',
       '_validation_status', '_notes', '_status', '_submitted_by',
       '__version__', '_tags', '_index','Cancer_screening_period(years)','routine_checkup_period(years)']
Data=Data.drop(columns=columns_to_drop)
```

```python
Data.columns
```

```python
# Function to clean text columns:
def clean_text_columns(df,columns):
    text_columns=df.select_dtypes(include='object').columns
    for cols in text_columns:
        df[cols]=df[cols].str.strip().str.lower()
    return df
Data=clean_text_columns(df=Data,columns=Data.columns)
print("Text columns cleaned successfully!")
```

```python
# Missing values and datatypes
print("Data types:\n",Data.dtypes)
Data.isnull().sum()
```

```python
# parsing date columns
Data['Date and Time']=pd.to_datetime(Data['Date and Time'], errors='coerce')

# Handling missing values in date columns
Data['Date and Time']=Data['Date and Time'].bfill()
Data['Date and Time']=Data['Date and Time'].dt.date

# fillna in gender , marital status and employment status
Data[['Gender','Age','Marital Status','Employment Status','Monthly Household Income']]=\
    Data[['Gender','Age','Marital Status','Employment Status','Monthly Household Income']].fillna('Not Provided')
```

```python
# cleaning Insurance columns
# A function to clean insurance cover column
Data['Had_Insurance ']=Data['Had_Insurance'].fillna('unknown')

def clean_insurance_cover(row):
    Have_insurance = str(row['Had_Insurance']).strip().lower()
    Insurance_company = row['Insurance_Company']  # Leave as-is for pd.isna()

    if Have_insurance == 'unknown':
        return 'Unknown'
    elif Have_insurance == 'no':
        return 'No Insurance'
    elif Have_insurance == 'yes':
        if pd.isna(Insurance_company) or str(Insurance_company).strip().lower() in ['nan', '', 'none', 'not provided']:
            return 'Not Specified'
        else:
            return Insurance_company
    else:
        return np.nan  # fallback for unexpected values
Data['Insurance_Company'] = Data.apply(clean_insurance_cover, axis=1)
```

```python
# flag to mark which row children are missing
Data['Chldren_Missing']=Data['Children'].isnull().astype(int)
# Group by Age range and fill with median number of children per group
Data['Children']=Data.groupby('Age')['Children'].transform(lambda x:x.fillna(x.median() if not np.isnan(x.median()) else Data['Children'].median()))
```

```python
# Handling Categorical columns
categorical_cols=['Had_Insurance','Last_time_visited_hospital(months)',
                  'Had_insurance_in_hospital','Has_checkup_with_doctor','Cancer_screening']
for cols in categorical_cols:
    Data[cols].fillna(Data[cols].mode()[0],inplace=True)

# Handling numerical location cols
Data.dropna(subset=['Location', '_Location_latitude', '_Location_longitude',
       '_Location_altitude', '_Location_precision'],inplace=True)

```
