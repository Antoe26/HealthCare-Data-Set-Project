```python
def percentage_finder(column_data):
    counts = column_data.value_counts(dropna=True)
    percentages = round((counts / counts.sum()) * 100, 2)
    return pd.DataFrame({'Count': counts, 'Percentage': percentages})
```

```python
# A function Generating countplots for all columns
def generate_countplots(Data,columns):
    for column in columns:
        # top 10
        top_n=10
        top_categories=Data[column].value_counts().nlargest(top_n).index
        filtered_data=Data[Data[column].isin(top_categories)]
        plt.figure(figsize=(8,5))
        sns.set(style='darkgrid')
        ax= sns.countplot(x=column, data=filtered_data, hue=column, palette='Set2',legend=False)
        plt.title(f"Count plot of {column}")
        plt.xticks(rotation=75)
        plt.tight_layout()
        plt.show()
        print(percentage_finder(Data[column]))
columns=[ 'Age','Gender', 'Marital Status', 'Children', 'Employment Status',
       'Monthly Household Income', 'Had_Insurance', 'Insurance_Company',
       'Last_time_visited_hospital(months)', 'Had_insurance_in_hospital',
       'Has_checkup_with_doctor', 'Cancer_screening']
generate_countplots(Data,columns)
```
